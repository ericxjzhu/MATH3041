# Analysing Data


Here we'll go through some useful techniques and methods used to analyse data.




Firstly, we can see that palm oil production is in millions, therefore to better understand the trend, we can plot the logarithm of the production. In python this is easily done using the `plt.semilogy` or `plt.semilogx` which plots log values of the y or x axis respectively.

## Method of Least Squares

The plot above gives what looks like linear trends (at least for Indonesia and the World). We can use the method of least squares to determine a line of best fit. For a linear equation of the form $\hat{y}=mx+b$ we have the following equations to solve for $ m $ and $b$, where $n$ is the number of data points,


\begin{align}
m &= \frac{n \sum x y - \left( \sum x \right) \left( \sum y\right) }{n \sum x^2 - \left( \sum x\right)^2}, \\
\\
b&= \frac{\sum y - m \sum x}{n}.
\end{align}

So that you can apply this to multiple curves to define a function. There are multiple ways to do this with other inbuilt python functions but we will do it from first principles.


````python
# Coomputing X and Y
def linear_fit(xx,yy):

  X = xx.values
  Y = yy.values

  # Total number of values
  n = len(X) 

  numer = n*sum(X*Y) - sum(X)*sum(Y)
  denom = n*sum(X**2) - sum(X)**2

  m = numer/denom

  b = (sum(Y)- m*sum(X))/n

  return(m, b) # return coefficents 
````

Using this definition we can apply it to the semi log data of Palm Oil production in Malaysia, Indonesia and the world.

````python
xval = np.linspace(1960,2030,1000) # define  values from 1960 to 2030 

# Determine constants for Indonesia
m_indonesia,b_indonesia = linear_fit(PO_indonesia[0],np.log(PO_indonesia[1]))
yval_PO_indo = m_indonesia*xval + b_indonesia # find y values 

# Least squares for Malaysia 
m_mal,b_mal = linear_fit(PO_malaysia[0],np.log(PO_malaysia[1]))
yval_PO_mal = m_mal*xval + b_mal

# Total palm oil production
m_world,b_world = linear_fit(years,np.log(PO_world))
yval_PO_world = m_world*xval + b_world

# define figure size
fig = plt.figure(figsize=(20,5))

# Plot all 3 linear fits side by side 
plt.subplot(1,3,1) # use plt.subplot to plot multiple figures
plt.scatter(PO_indonesia[0],np.log(PO_indonesia[1]),color='r',s=12)
plt.plot(xval,yval_PO_indo)
plt.title('Indonesia')
plt.legend(['Actual','Predicted'])
plt.xlabel('Years')
plt.ylabel('log Production (log(Tonnes))')

plt.subplot(1,3,2)
plt.scatter(PO_malaysia[0],np.log(PO_malaysia[1]),color='r',s=12)
plt.plot(xval,yval_PO_mal)
plt.title('Malaysia')
plt.legend(['Actual','Predicted'])
plt.xlabel('Years')
plt.ylabel('log Production (log(Tonnes))')

plt.subplot(1,3,3)
plt.scatter(years,np.log(PO_world),color='r',s=12)
plt.plot(xval,yval_PO_world)
plt.title('World')
plt.legend(['Actual','Predicted'])
plt.xlabel('Years')
plt.ylabel('log Production (log(Tonnes))')

````

By inspection, it looks like Indonesia and the world fits are reasonable. Malaysia's fit doesn't look quite right.

## Calculating R^2

To test the reliability of the curve we can determine the $R^2$ term, where the closer to one $R^2$ is the better the fit.
To calculate this, we use the following equations:



\begin{align}
SST &= \sqrt{\sum^n_{i=1}\left(y_i - \bar{y}\right)^2} \\
SSE &= \sqrt{\sum^n_{i=1}\left(y_i - \hat{y}_i\right)^2}\\
R^2 &= 1 - \frac{SSE}{SST}
\end{align}

We can now dbundle these formulas into a function.

````python
# Calculating R2 coefficient
def r_squared(xx,yy,m,c):

  X = xx.values
  Y = yy.values
  # Mean X and Y
  mean_x = np.mean(X)
  mean_y = np.mean(Y)
  # Total number of values
  n = len(X)

  ss_tot=0
  ss_res=0

  for i in range(n):
      y_pred = c + m * X[i]
      ss_tot += (Y[i] - mean_y) ** 2
      ss_res += (Y[i] - y_pred) ** 2
  r2 = 1 - (ss_res/ss_tot)
  return(r2)
````



Appling this function to the Palm Oil data sets:

````python
print('Indonesia')
print(r_squared(PO_indonesia[0],np.log(PO_indonesia[1]),m_indonesia,b_indonesia))
print('Malaysia')
print(r_squared(PO_malaysia[0],np.log(PO_malaysia[1]),m_mal,b_mal))
print('World')
print(r_squared(years,np.log(PO_world),m_world,c_world))
````

The fits for Indonesia and the world are fairly good, but Malaysia is not. This suggests that it's not an exponential function.


# Correlation to GDP

We might expect that palm oil production will be related to the gdp of both Indonesia and Malaysia, how much this has an impact will be determined by other economic activities of each country. To see how much palm oil production is related to gdp we can determine the correlation between the two datasets.

````python
url_gdp= 'https://raw.githubusercontent.com/janzika/MATH3041/main/data/gdp-per-capita-maddison-2020.csv'
gdp_data = pd.read_csv(url_gdp)

def gdp(country):
  cty = np.where(gdp_data['Entity']==country)[0]
  gdp_pc=gdp_data['GDP per capita'][cty[0]:cty[-1]]
  yrs = gdp_data['Year'][cty[0]:cty[-1]]

  return([yrs,gdp_pc])
````

````python
gdp_world = gdp('World')
gdp_indonesia = gdp('Indonesia')
gdp_malaysia = gdp('Malaysia')
````

Data for Palm Oil production starts at 1961, so that we can compare datasets we'll only consider gdp from this year.

````python
yrs_i = np.where(gdp('Indonesia')[0]>=1961)[0] # rows where year is greater than or equal to 1961
gdp_indonesia_new =[ gdp_indonesia[0][min(yrs_i):max(yrs_i)],gdp_indonesia[1][min(yrs_i):max(yrs_i)]] # new gdp data set

yrs_m = np.where(gdp('Malaysia')[0]>=1961)[0] # rows where year is greater than or equal to 1961
gdp_malaysia_new= [gdp_malaysia[0][min(yrs_m):max(yrs_m)],gdp_malaysia[1][min(yrs_m):max(yrs_m)]]
````

````python
fig = plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.plot(gdp_indonesia_new[0],gdp_indonesia_new[1])
plt.plot(gdp_malaysia_new[0],gdp_malaysia_new[1])
plt.legend(['Indonesia','Malaysia'])
plt.title('GDP')
plt.xlabel('Year')
plt.ylabel('GDP per capita (USD)')

plt.subplot(1,2,2)
plt.semilogy(gdp_indonesia_new[0],gdp_indonesia_new[1])
plt.semilogy(gdp_malaysia_new[0],gdp_malaysia_new[1])
plt.legend(['Indonesia','Malaysia'])
plt.title('Semi Log Plot of GDP')
plt.xlabel('Year')
plt.ylabel('GDP per capita (log(USD))')
````

````python

fig, ax = plt.subplots()

lns1=ax.plot(gdp_indonesia_new[0],gdp_indonesia_new[1],color='g',label='Indonesia GDP')
lns2=ax.plot(gdp_malaysia_new[0],gdp_malaysia_new[1],color='r',label='Malaysia GDP')
ax.set_ylabel('GDP per capitia (USD)')
ax.set_xlabel('Year')

# twin object for two different y-axis on the sample plot
ax2=ax.twinx()
# make a plot with different y-axis using second axis object
lns3=ax2.plot(PO_malaysia[0],PO_malaysia[1],label='Indonesia Palm Oil Production')
lns4=ax2.plot(PO_indonesia[0],PO_indonesia[1],label='Malaysia Palm Oil Production')
ax2.set_ylim([0,2.5e7])
ax2.set_ylabel('Palm Oil Production (Tonnes)')

# added these three lines
lns = lns1+lns3+lns2+lns4
labs = [l.get_label() for l in lns]
ax.legend(lns, labs, loc=0)
#fig.legend(['Indonesia','Malaysia','Indo palm oil','mal palm oil'],loc='upper left')

````
